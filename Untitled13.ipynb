```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("daily_temperature.csv")

# Plot line chart
plt.figure(figsize=(10,6))
plt.plot(df["Day"], df["Temperature"], marker='o', linestyle='-', color='b', label="Temperature")

# Add labels and title
plt.title("Daily Temperature Changes Over a Month")
plt.xlabel("Day of the Month")
plt.ylabel("Temperature (°C)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

# Show plot
plt.show()

# Interpretation: Find hottest and coldest days
hottest_day = df.loc[df["Temperature"].idxmax()]
coldest_day = df.loc[df["Temperature"].idxmin()]

print(f"🔥 Hottest Day: Day {hottest_day['Day']} with {hottest_day['Temperature']}°C")
print(f"❄️ Coldest Day: Day {coldest_day['Day']} with {coldest_day['Temperature']}°C")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("sales_data.csv")

# Plot line chart
plt.figure(figsize=(8,5))
plt.plot(df["Quarter"], df["Sales"], marker='o', linestyle='-', color='blue', label="Sales")

# Highlight peak sales with red marker
peak = df.loc[df["Sales"].idxmax()]
plt.scatter(peak["Quarter"], peak["Sales"], color='red', s=100, label=f"Peak Sales: {peak['Sales']}")

# Add labels, title, and grid
plt.title("Quarterly Sales Performance")
plt.xlabel("Quarter")
plt.ylabel("Sales (in $)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

# Show chart
plt.show()

# Print interpretation
print(f"📈 Peak Sales: {peak['Quarter']} with ${peak['Sales']}")
```
