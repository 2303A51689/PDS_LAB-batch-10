```python
import pandas as pd

# 1. Create Products dataset
products = pd.DataFrame({
    "product_id": [101, 102, 103, 104, 105, 106],
    "product_name": ["Laptop", "Mobile", "Tablet", "Headphones", "Charger", "Camera"],
    "category": ["Electronics", "Electronics", "Electronics", "Accessories", "Accessories", "Electronics"],
    "price": [60000, 20000, 15000, 3000, 1000, 25000]
})

# 2. Create Sales dataset
sales = pd.DataFrame({
    "sale_id": [1, 2, 3, 4, 5, 6],
    "product_id": [101, 102, 103, 104, 107, 105],
    "quantity": [3, 5, 2, 10, 4, 7],
    "date": pd.to_datetime(["2025-09-01", "2025-09-01", "2025-09-02", "2025-09-02", "2025-09-03", "2025-09-04"])
})

print("Products Dataset:\n", products, "\n")
print("Sales Dataset:\n", sales, "\n")

# 3. Merge datasets (Inner Join)
inner_merged = pd.merge(sales, products, on="product_id", how="inner")
print("Inner Join:\n", inner_merged, "\n")

# 4. Merge datasets (Outer Join)
outer_merged = pd.merge(sales, products, on="product_id", how="outer")
print("Outer Join:\n", outer_merged, "\n")

# 5. Compute statistics
inner_merged["revenue"] = inner_merged["quantity"] * inner_merged["price"]

total_revenue = inner_merged["revenue"].sum()
avg_price = inner_merged["price"].mean()

product_sales = inner_merged.groupby("product_name").agg(
    total_quantity=("quantity", "sum"),
    total_revenue=("revenue", "sum")
).reset_index()

category_sales = inner_merged.groupby("category").agg(
    total_quantity=("quantity", "sum"),
    total_revenue=("revenue", "sum")
).reset_index()

best_category = category_sales.loc[category_sales["total_revenue"].idxmax()]

print("Total Revenue:", total_revenue)
print("Average Price of Sold Products:", avg_price, "\n")
print("Product-wise Sales:\n", product_sales, "\n")
print("Category-wise Sales:\n", category_sales, "\n")
print("Best-Selling Category:\n", best_category, "\n")
```

```python
import pandas as pd

# 1. Create Teachers dataset
teachers = pd.DataFrame({
    "teacher_id": [1, 2, 3, 4, 5, 6],
    "teacher_name": ["Anita", "Rahul", "Priya", "Vivek", "Sonia", "Kiran"],
    "department": ["CSE", "ECE", "CSE", "MECH", "ECE", "CSE"]
})

# 2. Create Courses dataset
courses = pd.DataFrame({
    "course_id": [101, 102, 103, 104, 105, 106],
    "course_name": ["DBMS", "Networks", "AI", "Thermodynamics", "Signals", "OS"],
    "teacher_id": [1, 2, 1, 4, 2, 3],
    "credits": [4, 3, 3, 4, 3, 4]
})

print("Teachers Dataset:\n", teachers, "\n")
print("Courses Dataset:\n", courses, "\n")

# 3. Merge datasets (Inner Join)
inner_merged = pd.merge(courses, teachers, on="teacher_id", how="inner")
print("Inner Join (Teachers with assigned courses):\n", inner_merged, "\n")

# 4. Merge datasets (Left Join)
left_merged = pd.merge(teachers, courses, on="teacher_id", how="left")
print("Left Join (All teachers, even without courses):\n", left_merged, "\n")

# 5. Compute statistics

# Number of courses per teacher
courses_per_teacher = courses.groupby("teacher_id").size().reset_index(name="num_courses")

# Merge to get teacher names
courses_per_teacher = pd.merge(courses_per_teacher, teachers, on="teacher_id", how="right").fillna(0)

# Average courses per teacher
avg_courses = courses_per_teacher["num_courses"].mean()

# Department-wise workload (total courses per department)
dept_workload = pd.merge(courses, teachers, on="teacher_id", how="inner").groupby("department").agg(
    total_courses=("course_id", "count"),
    total_credits=("credits", "sum")
).reset_index()

# 6. Display results
print("Courses per Teacher:\n", courses_per_teacher, "\n")
print("Average Courses per Teacher:", avg_courses, "\n")
print("Department-wise Workload:\n", dept_workload, "\n")

# Best workload department (by total credits)
best_department = dept_workload.loc[dept_workload["total_credits"].idxmax()]
print("Department with Highest Workload:\n", best_department, "\n")
```
